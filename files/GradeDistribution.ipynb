```python
"""
https://www.pythoncharts.com/matplotlib/pie-chart-matplotlib/
"""
import matplotlib.pyplot as plt
prog = [22.5, 40, 37.5]
labels = ['GR', 'Labs', 'Homework']

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))
patches, texts, pcts = ax1.pie(prog, labels=labels, autopct='%.1f%%',
       wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       startangle=45,
       explode=(0, 0.1, 0),
       textprops={'size': 'x-large'})
for i, patch in enumerate(patches):
    texts[i].set_color(patch.get_facecolor())
plt.setp(pcts, color='white')
plt.setp(texts, fontweight=600)
ax1.set_title('Prog', fontsize=18)
plt.tight_layout()


final = [18, 32, 25, 25]
labels = ['GRs', 'Labs', 'Final Project', 'Homework']

patches, texts, pcts = ax2.pie(final, labels=labels, autopct='%.1f%%',
       wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       startangle=45,
       explode=(0, 0.1, 0, 0),
       textprops={'size': 'x-large'})
for i, patch in enumerate(patches):
    texts[i].set_color(patch.get_facecolor())
plt.setp(pcts, color='white')
plt.setp(texts, fontweight=600)
ax2.set_title('Final', fontsize=18)
plt.tight_layout()
plt.savefig('GradeDistribution.png')
```
